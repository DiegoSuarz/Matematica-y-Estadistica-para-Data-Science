##  

```python
# Definimos un conjunto de vectores en R^3
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v3 = np.array([7, 10, 15])

# Calculamos el determinante de la matriz formada por los vectores
matrix = np.array([v1, v2, v3])
matrix
```

```python
# Para saber si los vectores son linealmente independientes, quiere decir que al combinar dos vectore nos da resultado el que sobra
# Definimos un conjunto de vectores en R^3
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v3 = np.array([7, 10, 15])

# Calculamos el determinante de la matriz formada por los vectores
matrix = np.array([v1, v2, v3])
det = np.linalg.det(matrix) #sacar determinante

# Si el determinante es diferente de cero, los vectores son independientes linealmente
if det != 0:
    print("Los vectores son linealmente independientes.")
else:
    print("Los vectores son linealmente dependientes.")
```

```python
# Definición de vectores
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v3 = np.array([7, 8, 9])

# Comprobación de independencia lineal
matrix = np.vstack((v1, v2, v3))
matrix
```

```python
# Definición de vectores
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v3 = np.array([7, 8, 9])

# Comprobación de independencia lineal
matrix = np.vstack((v1, v2, v3))
det = np.linalg.det(matrix)
if det != 0:
    print("Los vectores son linealmente independientes.")
else:
    print("Los vectores son linealmente dependientes.")
```

```python
# Define los vectores como columnas de una matriz
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v3 = np.array([7, 10, 15])

# Crea la matriz con los vectores como columnas
matrix = np.column_stack((v1, v2, v3))

# Calcula el determinante de la matriz
det = np.linalg.det(matrix)

# Verifica si el determinante es distinto de cero
if det != 0:
    print("El conjunto de vectores es linealmente independiente.")
else:
    print("El conjunto de vectores es linealmente dependiente.")
```

##  

```python
# Definición de un subespacio vectorial
# Un subespacio vectorial es un conjunto de vectores que cumple con las propiedades de un espacio vectorial

# Por ejemplo, el conjunto de todos los vectores en R^2 que tienen la coordenada y igual a cero forma un subespacio vectorial.
# Podemos definirlo como el conjunto de todos los vectores (x, y) donde y = 0.
# El R2 es un subespacio de R3
```

```python
# Definimos un espacio vectorial
V = np.array([[1, 2, 3], [4, 5, 6], [7, 10, 15]])

# Definimos un subespacio vectorial
W = np.array([[1, 2, 3], [4, 5, 6]])

# Verificamos si W es un subespacio vectorial de V
if np.all(np.isin(W, V)):
    print("W es un subespacio vectorial de V")
else:
    print("W no es un subespacio vectorial de V")
```

```python
# Definición de una función de transformación lineal en R^2
def linear_transformation(vector):
    # Rotación de 90 grados en sentido antihorario
    transformation_matrix = np.array([[0, -1], [1, 0]])
    result = np.dot(transformation_matrix, vector)
    return result

# Aplicación de la transformación a un vector específico
v = np.array([1, 0])
result = linear_transformation(v)
print("Resultado de la transformación lineal:", result)
```

<a name="id7"></a> <h2> Matrices de transformación </h2>
##  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotacion(angulo):
  """
  Matriz de transformación para una rotación en el plano.

  Parámetros:
    angulo: El ángulo de rotación en radianes.

  Retorno:
    Matriz de transformación 2x2.
  """
  ca = np.cos(angulo)
  sa = np.sin(angulo)
  return np.array([[ca, -sa], [sa, ca]])

def escalamiento(escala):
  """
  Matriz de transformación para un escalamiento uniforme en el plano.

  Parámetros:
    escala: El factor de escala.

  Retorno:
    Matriz de transformación 2x2.
  """
  return np.array([[escala, 0], [0, escala]])

def reflexion_x():
  """
  Matriz de transformación para una reflexión respecto al eje x.

  Retorno:
    Matriz de transformación 2x2.
  """
  return np.array([[-1, 0], [0, 1]])

def graficar_vectores(vector_original, vector_transformado, titulo):
  """
  Grafica el vector original y el vector transformado.

  Parámetros:
    vector_original: El vector original.
    vector_transformado: El vector transformado.
    titulo: El título del gráfico.
  """
  # Líneas
  plt.plot([0, vector_original[0]], [0, vector_original[1]], color='black', linestyle='dashed', label='Vector original')
  plt.plot([0, vector_transformado[0]], [0, vector_transformado[1]], color='blue', label='Vector transformado')
  # Puntos
  plt.scatter([0, vector_original[0]], [0, vector_original[1]], color='black', marker='o')
  plt.scatter([0, vector_transformado[0]], [0, vector_transformado[1]], color='blue', marker='o')
  # Etiquetas
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(titulo)
  plt.legend()
  plt.show()
```

```python
# Rotación
angulo = np.pi / 4
matriz_rotacion = rotacion(angulo)
vector = np.array([1, 2])
vector_rotado = np.dot(matriz_rotacion, vector)

graficar_vectores(vector, vector_rotado, "Rotación en el plano")
```

```python
# Escalamiento
escala = -0.5
matriz_escalamiento = escalamiento(escala)
vector = np.array([1, 2])
vector_escalado = np.dot(matriz_escalamiento, vector)
print(vector_escalado)
graficar_vectores(vector, vector_escalado, "Escalamiento uniforme en el plano")
```

```python
# Reflexión
matriz_reflexion = reflexion_x()
vector = np.array([1, 2])
vector_reflejado = np.dot(matriz_reflexion, vector)
print(vector_reflejado)
graficar_vectores(vector, vector_reflejado, "Reflexión respecto al eje y")
```

```python
def plot_vector2d(vector2d, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)
```

```python
def plot_transformation(P_before, P_after, text_before, text_after, axis = [0, 5, 0, 4], arrows=False):
    if arrows:
        for vector_before, vector_after in zip(P_before.T, P_after.T):
            plot_vector2d(vector_before, color="blue", linestyle="--")
            plot_vector2d(vector_after, color="red", linestyle="-")
    plt.gca().add_artist(Polygon(P_before.T, alpha=0.2))
    plt.gca().add_artist(Polygon(P_after.T, alpha=0.3, color="r"))
    plt.text(P_before[0].mean(), P_before[1].mean(), text_before, fontsize=18, color="blue")
    plt.text(P_after[0].mean(), P_after[1].mean(), text_after, fontsize=18, color="red")
    plt.axis(axis)
    plt.grid()
```

```python
# La primera fila es igual a V1,∗P , que son las coordenadas de la proyección de P sobre el eje de 30°.
# La segunda fila es V2,∗P , que son las coordenadas de la proyección de P sobre el eje de 120°.
# Básicamente, obtuvimos las coordenadas de P después de rotar los ejes horizontal y vertical 30° (o equivalentemente después de rotar el polígono -30° alrededor del origen).

angle30 = 30 * np.pi / 180  # angle in radians
angle120 = 120 * np.pi / 180
V = np.array([
        [np.cos(angle30), np.sin(angle30)],
        [np.cos(angle120), np.sin(angle120)]
    ])
V
```

```python
# transformación conocida como mapeo de corte:
F_shear = np.array([
        [1, 1.5],
        [0, 1]
    ])
plot_transformation(P, F_shear.dot(P), "$P$", "$F_{shear} P$",
                    axis=[0, 10, 0, 7])
plt.show()
```

```python
# Definir el vector original
# v = np.array([0, 1])  # Vector unitario en dirección x
v = np.array([0.5, 0.5])

# Definir el ángulo de rotación en radianes
# theta = np.pi / 4  # Rotar 45 grados
theta = np.pi / 2

# Matriz de rotación en sentido antihorario
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

# Aplicar la rotación al vector original
v_rotated = np.dot(rotation_matrix, v)

print("El vector rotado es:", v_rotated)
```

```python
plot_vector2d(v, color="r")
plot_vector2d(v_rotated, color="b")
plt.axis([-1, 1, -1, 1])
plt.grid()
plt.show()
```

```python
# se genera 100 puntos igualmente espaciados entre 0 y 2pi
theta = np.linspace(0,2*np.pi,100)
# se generan puntos de coordenadas polares usando transformaciones seno y conseno
points = np.vstack((np.sin(theta),np.cos(theta)))
fig,ax = plt.subplots(1,figsize=(6,6))
# plth, = ax.plot(np.cos(x),np.sin(x),'ko')
plth, = ax.plot(np.cos(theta),np.sin(theta),'ko')
```

```python
def aframe(ph):
  # create and apply the transform matrix

  # creamos la matriz de transformación
  #  1° columna se desplaza a la derecha (1 unidad) y 2° columna se desplaza hacia arriba (1-ph)
  #  la segunda fila permanece sin cambio (no hay cambios en la coordenada "y")
  T = np.array([[1,1-ph],[0,1]])

  # aplicamos la matriz T a las coordenadas generadas
  P = T@points
  # update the dots' location
  plth.set_xdata(P[0,:])
  plth.set_ydata(P[1,:])
  return plth
```

```python
from matplotlib.animation import FuncAnimation
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
# generamos nuevos elementos para controlar la transformación a lo largo del parámetro ph
phi = np.linspace(-1,1-1/40,40)**2
anim=FuncAnimation(fig, aframe, phi, interval=100, repeat=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage.transform import warp

# Cargar una imagen de ejemplo (por ejemplo, la luna)
image = data.moon()

# Definir la matriz de transformación afín
matrix = np.array([[np.cos(np.pi/6), -np.sin(np.pi/6), 0],
                   [np.sin(np.pi/6), np.cos(np.pi/6), 0],
                   [0, 0, 1]])

# Aplicar la transformación afín a la imagen
transformed_image = warp(image, matrix)

# Visualizar la imagen original y la transformada
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Imagen Original')
axes[0].axis('off')
axes[1].imshow(transformed_image, cmap='gray')
axes[1].set_title('Imagen Transformada')
axes[1].axis('off')
plt.show()
```
