<a name="id2"></a> <h2>Estructura condicionale y Bucles </h2>
##  

### Estructura condicional simple y compuesta
Una sentencia condicional consiste básicamente en las siguientes partes principales:

- una prueba que evalúa a verdadero o falso
- un bloque de código que se ejecuta si la prueba es verdadero
- un bloque opcional de código si la prueba es falsa

```python
# Estructura condicional simple
x = 10
if x > 0:
    print("x es positivo")
else:
    print("x es negativo")
```

```python
# Estructura condicional compuesta
x = 0
if x > 0:
    print("x es positivo")
elif x == 0:
    print("x es cero")
else:
    print("x es negativo")
```

### Bucles

Los bucles son una estructura fundamental en la programación, que nos permiten ejecutar un bloque de código repetidamente mientras se cumple una condición. En Python, existen dos tipos de bucles: "for" y "while".

```python
# Bucle for
for i in range(5):
    print(i)
```

```python
# Bucle while
contador = 0
while contador < 5:
    print(contador)
    contador += 1
```

<a name="id3"></a> <h2>Funciones </h2>
##  

En Python, puedes definir tus propias funciones utilizando la palabra clave def.

```python
def suma(a, b):
    return a + b

resultado = suma(3, 5)
print(resultado)  # Salida: 8
```

<a name="id6"></a> <h2>SciPy </h2>
##  

Scipy es una biblioteca de Python de código abierto que se utiliza para realizar operaciones matemáticas, científicas y de ingeniería. Proporciona una amplia gama de funcionalidades para realizar cálculos numéricos eficientes, manipulación de datos, optimización, álgebra lineal, estadísticas y más.

 Operaciones de Álgebra Lineal

```python
import numpy as np
from scipy.linalg import solve, qr

# Crear una matriz A y un vector b
A = np.array([[2, 1], [1, 3]])
b = np.array([4, 5])

# Resolver el sistema de ecuaciones lineales Ax = b
x = solve(A, b)
print("Solución del sistema de ecuaciones lineales:", x)
```

Optimización Numérica

```python
from scipy.optimize import minimize

# Definir una función objetivo
def objetivo(x):
    return (x[0] - 1)**2 + (x[1] - 2)**2

# Realizar la minimización de la función objetivo
resultado = minimize(objetivo, [0, 0])
print("Mínimo de la función:", resultado.x)
```

 Integración Numérica

```python
from scipy.integrate import quad

# Definir una función a integrar
def f(x):
    return np.exp(-x**2)

# Realizar la integración numérica de la función
resultado, error = quad(f, 0, np.inf)
print("Resultado de la integración:", resultado)
print("Error estimado:", error)
```

Estadísticas

```python
from scipy.stats import norm

# Generar números aleatorios distribuidos normalmente
datos_aleatorios = norm.rvs(loc=0, scale=1, size=1000)
print("Datos aleatorios distribuidos normalmente:")
print(datos_aleatorios[:10])  # Mostrar los primeros 10 datos
```

<a name="id7"></a> <h2>Matplotlib </h2>
##  

Matplotlib es una biblioteca de visualización de datos en Python que permite crear gráficos de alta calidad y personalizables. En este tutorial, aprenderás cómo utilizar Matplotlib para crear diferentes tipos de gráficos, como gráficos de líneas, de dispersión, de barras, entre otros.

```python
import matplotlib.pyplot as plt
```

 Gráfico de Líneas

```python
# Datos
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear gráfico de líneas
plt.plot(x, y)

# Mostrar gráfico
plt.show()
```

Gráfico de Dispersión

```python
# Datos
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear gráfico de dispersión
plt.scatter(x, y)

# Mostrar gráfico
plt.show()
```

Gráfico de Barras

```python
# Datos
x = ['A', 'B', 'C', 'D', 'E']
y = [20, 35, 30, 25, 40]

# Crear gráfico de barras
plt.bar(x, y)

# Mostrar gráfico
plt.show()
```

Personalización de Gráficos

```python
# Datos
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 9, 16, 25]

# Crear gráfico de líneas con personalización
plt.plot(x, y1, label='Primos', color='blue', linestyle='--', marker='o')
plt.plot(x, y2, label='Cuadrados', color='red', linestyle='-', marker='x')

# Añadir etiquetas y leyenda
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Líneas')
plt.legend()

# Mostrar cuadrícula
plt.grid(True)

# Mostrar gráfico
plt.show()
```
