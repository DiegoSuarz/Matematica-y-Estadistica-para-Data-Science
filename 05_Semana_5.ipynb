```python
import sympy as sp

def es_continua_en_punto(f, x):
  """
  Verifica si una función es continua en un punto x.

  Args:
    f: Función a evaluar.
    x: Punto en el que se evalúa la continuidad.

  Devuelve:
    True si la función es continua en x, False en caso contrario.
  """
  try:
    # Se evalúa el valor de la función en el punto.
    valor_funcion = f(x)

    # Se evalúan los límites unilaterales derecho e izquierdo.
    limite_derecha = sp.limit(f(x), x, x + 0.0001, dir='+')
    limite_izquierda = sp.limit(f(x), x, x - 0.0001, dir='-')

    # Se verifica si los límites y el valor de la función coinciden.
    if limite_derecha == valor_funcion == limite_izquierda:
      return True
    else:
      return False
  except:
    # Si hay una excepción al evaluar la función o los límites, se considera que la función no es continua.
    return False

# def es_continua_en_punto(f, x):
#   """
#   Verifica si una función es continua en un punto x.

#   Args:
#     f: Función simbólica a evaluar (definida con sp.symbols).
#     x: Punto en el que se evalúa la continuidad.

#   Devuelve:
#     True si la función es continua en x, False en caso contrario.
#   """
#   try:
#     # Se evalúa el valor de la función en el punto (simbólicamente).
#     valor_funcion = f.subs(x, x)

#     # Se evalúan los límites unilaterales derecho e izquierdo usando sp.limit.
#     limite_derecha = sp.limit(f, x, x + 0.0001, dir='+')
#     limite_izquierda = sp.limit(f, x, x - 0.0001, dir='-')

#     # Se verifica si los límites y el valor de la función coinciden.
#     if limite_derecha == valor_funcion == limite_izquierda:
#       return True
#     else:
#       return False
#   except:
#     # Si hay una excepción al evaluar la función o los límites, se considera que la función no es continua.
#     return False

def es_continua_en_intervalo(f, a, b):
  """
  Verifica si una función es continua en un intervalo [a, b].

  Args:
    f: Función a evaluar.
    a: Extremo inferior del intervalo.
    b: Extremo superior del intervalo.

  Devuelve:
    True si la función es continua en el intervalo, False en caso contrario.
  """
  # Se verifica la continuidad en cada punto del intervalo con una discretización fina.
  for x in np.arange(a, b + 0.01, 0.01):
    if not es_continua_en_punto(f, x):
      return False

  # Si todas las pruebas de continuidad en puntos discretos son positivas, se considera que la función es continua en el intervalo.
  return True
```

```python
import sympy as sp

x = sp.symbols('x')  # Define symbolic variable 'x'
f = x**2  # Define the function f(x) = x^2

# Calculate the symbolic derivative
df_dx = f.diff(x) #diff permirte aplicar la derivada
print("Derivada simbólica de f(x) =", f)

print("Derivada de f(x): ",df_dx)
# Evaluate the derivative at a specific value
x_val = 3  # Evaluate at x = 3
numerical_derivative = df_dx.evalf(subs={x: x_val})
print("Valor numérico de la derivada en x =", x_val, ":", numerical_derivative)
```

```python
import sympy as sp

x = sp.symbols('x')  # Define symbolic variable 'x'

f = x**2  # Define function f(x)
g = sp.exp(x)  # Define function g(x)

# Product of functions
product_function = f * g # f'*g+f*g'

# Calculate derivative using product rule
df_dx = sp.diff(product_function, x)
print("Derivada del producto de funciones f(x) * g(x) usando la regla del producto:", df_dx)
```

```python
# Chain Rule (regla de la cadena)
import sympy as sp

x = sp.symbols('x')  # Define symbolic variable 'x'

# Define inner and outer functions
inner_function = x**2
outer_function = lambda y: sp.sin(y)

# Composite function
composite_function = outer_function(inner_function)

# Calculate derivative using chain rule
df_dx = sp.diff(composite_function, x)
print("Derivada de la función compuesta f(g(x)) usando la regla de la cadena:", df_dx)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Definición de la función
def f(x):
    return x**2 - 4*x + 4  # Ejemplo de función cuadrática

# Derivada de la función
def df(x):
    return 2*x - 4  # Derivada de la función cuadrática
    # return sp.diff(f, x)

# Descenso del gradiente univariado
def gradient_descent_univariate(learning_rate, initial_x, num_iterations):
    x_values = [initial_x]
    for _ in range(num_iterations):
        gradient = df(x_values[-1])
        new_x = x_values[-1] - learning_rate * gradient
        x_values.append(new_x)
    return x_values

# Parámetros
learning_rate = 0.1
initial_x = 4  # Valor inicial de x
num_iterations = 20

# Ejecutar el descenso del gradiente
x_values = gradient_descent_univariate(learning_rate, initial_x, num_iterations)

# Crear gráfico de la función y el descenso del gradiente
x_vals = np.linspace(-1, 5, 100)
y_vals = f(x_vals)

fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, label='f(x)')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Descenso del Gradiente Univariado')
line, = ax.plot([], [], 'ro', animated=True)

# Función de inicialización de la animación
def init():
    line.set_data([], [])
    return line,

# Función de actualización de la animación
def update(frame):
    x = x_values[frame]
    y = f(x)
    line.set_data(x, y)
    return line,

# Crear la animación
ani = animation.FuncAnimation(fig, update, frames=len(x_values), init_func=init, blit=True)
ani
plt.legend()
plt.grid()
plt.show()
```

```python
import sympy as sp

x, y = sp.symbols('x y')  # Define symbolic variables 'x' and 'y'

f = 2*x**2 + 3*x*y + y**3  # Define the function f(x, y)

# Calculate partial derivative with respect to x
df_dx = f.diff(x)
print("Derivada parcial de f(x, y) con respecto a x:", df_dx)

# Calculate partial derivative with respect to y
df_dy = f.diff(y)
print("Derivada parcial de f(x, y) con respecto a y:", df_dy)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función
def f(x, y):
    return 2*x**2 + 3*x*y + y**3

# Generación de puntos en el dominio
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Gráfico de la función
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')

# Título del gráfico
ax.set_title('Gráfico de f(x, y) = 2x^2 + 3xy + y^3')

# Mostrar el gráfico
plt.show()
```

```python
import sympy as sp

x, y, z = sp.symbols('x y z')  # Define symbolic variables 'x', 'y', and 'z'

f = x**2 + y**2 + z**3 - 2*x*y + 3*y*z - x*z  # Define the function f(x, y, z)

# Calculate partial derivative with respect to x
df_dx = f.diff(x)
print("Derivada parcial de f(x, y, z) con respecto a x:", df_dx)

# Calculate partial derivative with respect to y
df_dy = f.diff(y)
print("Derivada parcial de f(x, y, z) con respecto a y:", df_dy)

# Calculate partial derivative with respect to z
df_dz = f.diff(z)
print("Derivada parcial de f(x, y, z) con respecto a z:", df_dz)
```

```python
import sympy as sp

x, y = sp.symbols('x y')  # Define symbolic variables 'x' and 'y'

# Implicit equation: x^3 + y^2 = 2xy
equation = x**3 + y**2 - 2*x*y

# Differentiate both sides implicitly
equation_diff = sp.diff(equation, x)
print("Derivada respecto a 'x' de la ecuación x^3 + y^2 = 2xy:", equation_diff)

# Differentiate both sides implicitly
equation_diff = sp.diff(equation, y)
print("Derivada respecto a 'y' de la ecuación x^3 + y^2 = 2xy:", equation_diff)
```

```python
# ACKLEY FUNCTION
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir las variables simbólicas
x, y = sp.symbols('x y')

# Definir la función de costo simbólica
f = -20 * sp.exp(-0.2 * sp.sqrt((x**2 + y**2) / 2)) - sp.exp((sp.cos(2 * sp.pi * x) + sp.cos(2 * sp.pi * y)) / 2) + 20 + sp.exp(1)

# Calcular el gradiente simbólico
grad = [sp.diff(f, var) for var in (x, y)]

# Convertir la función de costo y el gradiente en funciones numéricas
cost_function_numeric = sp.lambdify((x, y), f)
gradient_numeric = sp.lambdify((x, y), grad)

# Definir el rango de valores para x e y
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)

# Crear una malla de puntos
X, Y = np.meshgrid(x_vals, y_vals)

# Evaluar la función de costo en la malla de puntos
Z = cost_function_numeric(X, Y)

# Definir los valores iniciales y el learning rate para el descenso del gradiente
initial_values = np.array([-4, -4])
learning_rate = 0.1
num_iterations = 50

# Realizar el descenso del gradiente y guardar los puntos de iteración
iteration_points = [initial_values.copy()]
for _ in range(num_iterations):
    gradient = gradient_numeric(*initial_values)
    initial_values = initial_values - learning_rate * np.multiply(gradient, np.ones_like(initial_values))
    iteration_points.append(initial_values.copy())

# Convertir los puntos de iteración en un arreglo numpy
iteration_points = np.array(iteration_points)

# Graficar la superficie y los puntos de iteración
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)
ax.scatter(iteration_points[:, 0], iteration_points[:, 1], cost_function_numeric(iteration_points[:, 0], iteration_points[:, 1]), color='r', label='Puntos de Iteración')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Costo')
ax.set_title('Superficie de la función de costo con Puntos de Iteración')
ax.legend()
ax.view_init(elev=5, azim=15)

# Graficar la superficie y los puntos de iteración
ax2 = fig.add_subplot(122)
ax2.contour(X, Y, Z, levels=20, cmap='viridis')
ax2.scatter(iteration_points[1:, 0], iteration_points[1:, 1], color='y', label='Puntos de Iteración')
ax2.scatter(iteration_points[0, 0], iteration_points[0, 1], color='b', label='Puntos de Inicial')
ax2.scatter(iteration_points[-1, 0], iteration_points[-1, 1], color='r', label='Punto de Iteración Final')
# ax.plot(iteration_points[:, 0], iteration_points[:, 1], color='b', linewidth=1, linestyle='--', label='Recorrido')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Contornos de la función de costo con Puntos de Iteración')
ax2.legend()

fig.tight_layout()
```

```python
# intialize animation
def init_fig():

    """Initialise figure"""
    fig = plt.figure(figsize=plt.figaspect(0.3))


    ax1=fig.add_subplot(1,3,3,projection='3d') #ax1 is the surface plot of loss
    tx,ty = np.meshgrid(t0,t1,indexing='ij')
    ax1.plot_wireframe(tx,ty,cost)
    #this would ideally be a surface plot rather than wireframe
    #but in matplotlib that makes the points hidden below the plane

    pc_dots, = ax1.plot([],[],[],'ro')
    cost_text = ax1.set_title('loss at step {}'.format(0), fontsize=18)
    ax1.set_xlabel(r'$\theta_0$')
    ax1.set_ylabel(r'$\theta_1$')
    ax1.set_zlabel('loss')
    ax1.zaxis.set_label_position('top')
    line1,=ax1.plot([],[],[],'r', label='descent path')

    ax12 = fig.add_subplot(1,3,2)    #ax12 is contour plot of loss
    contour = ax12.contour(tx,ty,cost,levels=25)
    pc_dotc,=ax12.plot([],[],'ro')
    ax12.set_xlabel(r'$\theta_0$')
    ax12.set_ylabel(r'$\theta_1$')
    line12, = ax12.plot([],[], 'r', label='descent path')

    ax2 = fig.add_subplot(1,3,1) #ax2 is the data with fit
    ax2.plot(X[:,1], Y, 'bo', label='data')
    ax2.set_xlabel('input: $x_1$', fontsize=15)
    ax2.set_ylabel('Y', fontsize=15)
    ax2.yaxis.set_label_position('left')
    ax2.set_title('Labelled data & model output', fontsize=18)
    line2, = ax2.plot([], [], 'k-', label='fitted line: $y=x\\theta$')
    ax2.legend(loc=2)
    return fig, ax1, ax2, line1, line12,line2, pc_dots, pc_dotc,cost_text


def get_anim(fig, ax1, ax2, line1, line12,line2, pc_dots, pc_dotc,cost_text, cost_list,theta_hist0, theta_hist1,X):
    def anim(i):
        """Animate step i"""
        if i == 0:
            return [line2, pc_dots, cost_text]

        thetai=[theta_hist0[i-1], theta_hist1[i-1]]
        cost_text.set_text('Loss at step {} = {:.3f}'.format(i, cost_list[i-1]))
    #surface plot
        pc_dots.set_xdata(theta_hist0[i-1])
        pc_dots.set_ydata(theta_hist1[i-1])
        pc_dots.set_3d_properties(cost_list[i-1])

    #contour plot
        pc_dotc.set_xdata(theta_hist0[i-1])
        pc_dotc.set_ydata(theta_hist1[i-1])

        line1.set_xdata(theta_hist0[:i])
        line1.set_ydata(theta_hist1[:i])
        line1.set_3d_properties(cost_list[:i])
        line12.set_xdata(theta_hist0[:i])
        line12.set_ydata(theta_hist1[:i])
        ax2.legend(loc=2)
    #data with fit
        xli=np.array([[1,-10],[1,10]])
        Y_pred=xli@thetai
        line2.set_xdata(xli[:,1])
        line2.set_ydata(Y_pred)

        return [line1, line12,line2, pc_dots,pc_dotc,cost_text]
    return anim
```

```python
from __future__ import division, print_function, absolute_import, unicode_literals

# Python imports
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

%matplotlib inline
np.random.seed(seed=1)

from matplotlib import animation, rc
from IPython.display import HTML, Image

rc('animation', html='html5')

#create a dataset with 100 examples,each containing two variables
m=100 #no. of examples
X=np.ones([m,2]) #adds X0 to the X values
#X[:,0]=np.linspace(-5,5,100)
X[:,1]=np.linspace(-5,5,m)
Y=np.ones(m)
theta_set=[2,2] #parameters for data generation (intercept and slope)
noise = np.random.randn(len(Y))*0.2
Y=X@theta_set+noise*10

# calculate general cost function shape for plot
no_dots = 40
t0 = np.linspace(-5,10, no_dots)
t1 = np.linspace(-1,5, no_dots)
thetag=[]
cost = np.zeros([no_dots,no_dots])

for i in range(no_dots):
    for j in range(no_dots):
        thetag=[t0[i],t1[j]]
        cost[i,j]=(1/(2*m))*np.transpose((X@thetag - Y))@(X@thetag - Y)

#GRADIENT DESCENT
theta = np.ones(2)
cost_list=[]
theta_hist0=[]
theta_hist1=[]


###### to be controlled by the user##############
theta[0]=0 #starting intercept value
theta[1]=5 #starting slope value
epochs = 30 #number of epochs
lr=0.2 #learning rate
##################################################

theta_hist0.append(theta[0])
theta_hist1.append(theta[1])

for i in range(epochs):
  theta = theta - lr*(1/m)*np.transpose(X)@(X@theta - Y)
  theta_hist0.append(theta[0])
  theta_hist1.append(theta[1])
  cost_val = (1/(2*m))*np.transpose((X@theta - Y))@(X@theta - Y)
  cost_list.append(cost_val)
```
